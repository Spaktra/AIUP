```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv("data/CC_GENERAL.csv")
dataset.shape
```

```python
dataset.head(10)
```

```python
X = dataset.drop(["CUST_ID"], axis = True)
```

```python
X.info()
```

```python
X.describe()
```

```python
X.isna().sum()
```

```python
X["CREDIT_LIMIT"] = X["CREDIT_LIMIT"].fillna(float(np.mean(X["CREDIT_LIMIT"])))
X["MINIMUM_PAYMENTS"] = X["MINIMUM_PAYMENTS"].fillna(float(np.mean(X["MINIMUM_PAYMENTS"])))
```

```python
X.isna().sum()
```

```python
labels = X.columns
labels[1]
for i in labels:
    print(i)
```

```python
labels = X.columns
for i in labels:
    plt.figure(figsize = (6,6))
    plt.hist(X[i], bins = 30, color = 'yellowgreen', edgecolor = '#6A9662')
    plt.gca().set(title = i, ylabel = 'Frekans')
    plt.show()
```

```python
plt.figure(figsize=(18, 18))
heatmap = sns.heatmap(X.corr(), annot=True, cmap = 'GnBu_r')
heatmap.set_title('Corelation Heatmap')
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

wcss=[]

for i in range(1,20):
    kmeans=KMeans(n_clusters=i, init="k-means++",random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    

plt.figure(figsize=(13,8))
plt.plot(range(1,20),wcss, marker = 'o')

plt.xticks(range(1,20))
plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=8,init="k-means++",random_state=0)
y_kmeans=kmeans.fit_predict(X)
```

```python
y_kmeans.max()
```

```python
plt.figure(figsize = (10,8))
plt.hist(y_kmeans, bins = 30, color = '#DF7861', edgecolor = '#94B49F')
plt.gca().set(title = "Clusters", ylabel = 'Frekans')
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(X)
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize = (13,8))
plt.plot(range(1,18), pca.explained_variance_ratio_.cumsum(), marker = 'o')
plt.title("Explained Varience by Components")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.show()
```

```python
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)
```

```python
X_pca
```

```python
y_kmeans_pca=kmeans.fit_predict(X_pca)
y_kmeans_pca.max()
```

```python
X_pca = pd.DataFrame(X_pca,columns=["Feature-1","Feature-2"])
X_pca.head()
```

```python
plt.scatter(X_pca[y_kmeans_pca==0]["Feature-1"],X_pca[y_kmeans_pca==0]["Feature-2"],label="cluster1")
plt.scatter(X_pca[y_kmeans_pca==1]["Feature-1"],X_pca[y_kmeans_pca==1]["Feature-2"],label="cluster2")
plt.scatter(X_pca[y_kmeans_pca==2]["Feature-1"],X_pca[y_kmeans_pca==2]["Feature-2"],label="cluster3")
plt.scatter(X_pca[y_kmeans_pca==3]["Feature-1"],X_pca[y_kmeans_pca==3]["Feature-2"],label="cluster4")
plt.scatter(X_pca[y_kmeans_pca==4]["Feature-1"],X_pca[y_kmeans_pca==4]["Feature-2"],label="cluster5")
plt.scatter(X_pca[y_kmeans_pca==5]["Feature-1"],X_pca[y_kmeans_pca==5]["Feature-2"],label="cluster6")
plt.scatter(X_pca[y_kmeans_pca==6]["Feature-1"],X_pca[y_kmeans_pca==6]["Feature-2"],label="cluster7")
plt.scatter(X_pca[y_kmeans_pca==7]["Feature-1"],X_pca[y_kmeans_pca==7]["Feature-2"],label="cluster8")


plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100, label="cluster center", c = 'black')
plt.title("Clusters of customers")
plt.legend()
plt.show()
```
